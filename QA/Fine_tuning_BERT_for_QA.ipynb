```python
import pandas as pd

df = pd.read_csv('/content/BankChurners.csv')
column_names = df.columns.tolist()
print(column_names)
```

```python
# Drop the last two columns since they are not useful for analysis
df.drop(df.columns[-2:], axis=1, inplace=True)

# Drop any rows with missing values
df.dropna(inplace=True)

# Remove any whitespace from column names
df.columns = df.columns.str.strip()

# Convert 'Income_Category' column to a categorical variable
df['Income_Category'] = pd.Categorical(df['Income_Category'], categories=['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +'], ordered=True)

# Convert 'Gender' column to a binary variable
df['Gender'] = pd.Series(df['Gender'] == 'M', dtype=int)

# Convert 'Attrition_Flag' column to a binary variable
df['Attrition_Flag'] = pd.Series(df['Attrition_Flag'] == 'Attrited Customer', dtype=int)

# Convert 'Education_Level' column to an ordered categorical variable
df['Education_Level'] = pd.Categorical(df['Education_Level'], categories=['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate'], ordered=True)

```

```python
df.head(5)
```

```python
# Count the number of null or empty values in each row
null_counts = df.isnull().sum(axis=1)

# Remove rows with more than 5 null or empty values
df = df[null_counts <= 5]
```

```python
df = df.drop('Attrition_Flag', axis=1)
```

```python
import pandas as pd

def preprocess_row(row):
    # Define the context
    context = f"Customer {row['CLIENTNUM']} is a {row['Customer_Age']}-year-old {row['Gender']} with {row['Education_Level']} education, {row['Dependent_count']} dependents, and {row['Income_Category']} income category. They have {row['Total_Relationship_Count']} total relationships with the bank, a credit limit of {row['Credit_Limit']} dollars, and a revolving balance of {row['Total_Revolving_Bal']} dollars."

    # Define a list of questions and answers
    qa_pairs = []

    # Add question and answer for total transactions amount
    question = f"How much money did Customer {row['CLIENTNUM']} spend on transactions?"
    answer = str(row['Total_Trans_Amt'])
    qa_pairs.append((question, answer))

    # Add question and answer for average utilization ratio
    question = f"What is the average utilization ratio for Customer {row['CLIENTNUM']}?"
    answer = str(row['Avg_Utilization_Ratio'])
    qa_pairs.append((question, answer))

    # Add question and answer for credit limit
    question = f"What is the credit limit for Customer {row['CLIENTNUM']}?"
    answer = str(row['Credit_Limit'])
    qa_pairs.append((question, answer))

    # Add question and answer for months inactive in last 12 months
    question = f"How many months was Customer {row['CLIENTNUM']} inactive in the last 12 months?"
    answer = str(row['Months_Inactive_12_mon'])
    qa_pairs.append((question, answer))

    # Create a list of lists representing the dataset rows
    dataset_rows = []
    for i, (question, answer) in enumerate(qa_pairs):
        # Define the answer text and start position
        try:
            answer_text = answer
            answer_start = context.index(answer_text)
        except ValueError:
            answer_text = ""
            answer_start = -1

        # Create a list representing the dataset row
        dataset_row = [f"{row['CLIENTNUM']}-{i}", context, question, answer_text, answer_start]

        dataset_rows.append(dataset_row)

    return dataset_rows


# Preprocess each row and flatten the resulting list of lists
qa_rows = [row for rows in df.apply(preprocess_row, axis=1) for row in rows]

# Write the output CSV file
with open('qa.csv', 'w') as f:
    f.write('id,context,question,answer_text,answer_start\n')
    for row in qa_rows:
        f.write(','.join(str(field) for field in row) + '\n')
```

```python
import pandas as pd
import torch
from transformers import BertForQuestionAnswering, BertTokenizer

# Load the QA CSV file into a pandas DataFrame
qa_df = pd.read_csv('/content/qa.csv')

# Initialize the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForQuestionAnswering.from_pretrained('bert-base-uncased')

# Set the device to GPU if available, otherwise use CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define a function to tokenize a row of data and return the input IDs and attention masks
def tokenize_row(row):
    encoding = tokenizer.encode_plus(row['question'], row['context'], max_length=512, pad_to_max_length=True, return_tensors='pt')
    input_ids = encoding['input_ids'].squeeze()
    attention_mask = encoding['attention_mask'].squeeze()
    return input_ids, attention_mask

# Tokenize the questions and contexts and add them to the DataFrame
qa_df['input_ids'], qa_df['attention_mask'] = zip(*qa_df.apply(tokenize_row, axis=1))

# Define the training function
def train(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0
    for batch in dataloader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        start_positions = batch['start_positions'].to(device)
        end_positions = batch['end_positions'].to(device)

        optimizer.zero_grad()

        loss, _, _ = model(input_ids, attention_mask=attention_mask, start_positions=start_positions, end_positions=end_positions)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(dataloader)

# Define the evaluation function
def evaluate(model, dataloader, device):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            start_positions = batch['start_positions'].to(device)
            end_positions = batch['end_positions'].to(device)

            loss, _, _ = model(input_ids, attention_mask=attention_mask, start_positions=start_positions, end_positions=end_positions)
            total_loss += loss.item()

    return total_loss / len(dataloader)

# Define the main function to train the model
def main():
    # Split the data into training and validation sets
    train_df = qa_df.sample(frac=0.8, random_state=42)
    val_df = qa_df.drop(train_df.index)

    # Create PyTorch datasets and dataloaders
    from torch.utils.data import Dataset, DataLoader

    class QADataset(Dataset):
        def __init__(self, df):
            self.df = df

        def __len__(self):
            return len(self.df)

        def __getitem__(self, index):
            row = self.df.iloc[index]
            input_ids = row['input_ids']
            attention_mask = row['attention_mask']
            start_position = row['answer_start']
            end_position = row['answer_start'] + len(row['answer_text'])
            return {'input_ids': input_ids, 'attention_mask': attention_mask, 'start_positions': start_position, 'end_positions': end_position}

    train_dataset = QADataset(train_df)
    val_dataset = QADataset(val_df)

    train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=8)

    # Define the optimizer and learning rate scheduler
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3, verbose=True)

    # Train the model for 3 epochs
    for epoch in range(4):
        train_loss = train(model, train_loader, optimizer, device)
        val_loss = evaluate(model, val_loader, device)
        scheduler.step(val_loss)

        print(f'Epoch {epoch + 1} train loss: {train_loss:.3f} | val loss: {val_loss:.3f}')

    # Save the fine-tuned model
    model.save_pretrained('fine_tuned_bert_qa')



    if name == 'main':
      main()
```

```python
import torch
from transformers import BertForQuestionAnswering, BertTokenizer

model_dir = '/content/fine_tuned_bert_qa/'

# Load the model and tokenizer
model = BertForQuestionAnswering.from_pretrained(model_dir)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Define a function to predict the answer given a question and a context
def predict_answer(question):
    # Tokenize the input question and context
    encoded_input = tokenizer.encode_plus(question, add_special_tokens=True, return_tensors='pt')
    input_ids = encoded_input['input_ids'].to(device)
    attention_mask = encoded_input['attention_mask'].to(device)

    # Generate predictions for the input
    start_scores, end_scores = model(input_ids, attention_mask=attention_mask)
    start_index = torch.argmax(start_scores)
    end_index = torch.argmax(end_scores)

    # Decode the predicted answer from the input IDs
    answer_tokens = input_ids[0][start_index:end_index+1]
    answer = tokenizer.decode(answer_tokens)

    return answer


# Test the function on a sample question
question = "What is the credit limit for Customer 708790833?"
answer = predict_answer(question)
print(answer)
```
